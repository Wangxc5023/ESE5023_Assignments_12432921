```python
def Find_number_of_ways(x):
  N_X=0
  for a in range(0,11):
      for b in range(0,11-a):
            for c in range(0,11-a-b):
                  for d in range(0,11-a-b-c):
                      for e in range(0,11-a-b-c-d):
                          for f in range(0,11-a-b-c-d-e):
                              if a+2*b+3*c+4*d+5*e+6*f==x and a+b+c+d+e+f==10:
                                 N_X=N_X+1
 
  print(N_X)
  return N_X
```

```python
def Sum_averages(N):
    random_integers = Random_integer(N)
    SUM_ = 0
    for Num_e in range(1, N+1): 
        SUM_ += math.comb(N-1, Num_e-1) / Num_e
    S_a = sum(random_integers) * SUM_
    print(S_a)
    return S_a
```

```python
Sum_averages(3)
```

```python
Total_sum_averages = []
for N in range(1, 101):
    total_sum_averages=Sum_averages(N)
    Total_sum_averages.append(total_sum_averages)
print(Total_sum_averages)
import matplotlib.pyplot as plt
plt.plot(Total_sum_averages)
plt.title('Total Sum Averages')
plt.xlabel('N')
plt.ylabel('S_a')
plt. show()
```

```python
def generate_matrix(N, M):
    # Without numpy
    arr = [[random.randint(0,1)] * (M) for _ in range(N)]
    arr[0][0]=1
    arr[N-1][M-1]=1

    # With numpy
    arr=np.random.randint(2,size=(N,M))
    arr[0,0]=1
    arr[N-1,M-1]=1
    print(arr)
    return arr
```

```python
# get right answer
def Count_path(M,N):
    matrix = generate_matrix(M, N)
    M, N = len(matrix), len(matrix[0])
    path_count = [0] # NOTE: 似乎不用定义成 list
    def dfs(i, j):
        if i >= M or j >= N or matrix[i][j] == 0:
            return 0
        if i == M - 1 and j == N - 1:
            path_count[0] += 1 # NOTE: 这里为什么 +1
            return
        temp = matrix[i][j]
        matrix[i][j] = 0
        dfs(i, j + 1)
        dfs(i + 1, j)
        matrix[i][j] = temp
    dfs(0, 0)
    return path_count[0]
```

```python
Count_path(5,5)
```

```python
SUM_COUNT=0
for NUM_COUNT in range(1,1001):
    SUM_COUNT+=Count_path(10,8)
AVE_COUNT=SUM_COUNT/1000
print(AVE_COUNT)
```
